### 문제 4. 튜플과 리스트의 결합

아래 두 튜플의 모든 요소를 하나의 리스트 `all_points`에 담아 `[10, 20, 30, 40]`이 출력되게 만드세요.

```python
point1 = (10, 20)
point2 = (30, 40)

set_point1 = set(point1)
set_point2 = set(point2)

all_point = set_point1 | set_point2

print(sorted(all_point))
```

````python
point1 = (10, 20)
point2 = (30, 40)

set_point1 = set(point1)
set_point2 = set(point2)

all_point = set_point1 | set_point2

print(sorted(all_point))
````

### 문제 5. 빈 데이터 판별기(심화)

사용자로부터 입력받은 검색어 `search_keyword`가 비어있는지 확인하는 코드를 작성하세요.

- 검색어가 있으면: **" '검색어'를 검색합니다."** 출력
- 검색어가 비어있거나 공백만 있다면(strip 활용): **"검색어를 입력해주세요."** 출력

```python
search_keyword = "  " # 예시: 공백만 입력된 경우

if search_keyword.strip() == "":
    print("검색어를 입력해주세요.")
else:
    print(f"{search_keyword}를 검색합니다.")
```

````python
search_keyword = "  "

if search_keyword.strip() == "":
    print("검색어를 입력해주세요.")
else:
    print(f"{search_keyword}를 검색합니다.")
````
